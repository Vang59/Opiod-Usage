```python
# Imports
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import scipy.stats as st
%matplotlib inline
```

```python
# Csv file
csv = ("Resources/master.csv")

# Create DataFrame
df = pd.read_csv(csv).set_index("State")
pd.set_option("display.max_columns", 22, "display.colheader_justify", "center")
df
```

```python
# Create DataFrame converting % to float
df["Usage: 18+ Estimate"]=df["Usage: 18+ Estimate"].str.rstrip("%").astype(float)/100
df["Usage: 18+ 95% CI (Lower)"]=df["Usage: 18+ 95% CI (Lower)"].str.rstrip("%").astype(float)/100
df["Usage: 18+ 95% CI (Upper)"]=df["Usage: 18+ 95% CI (Upper)"].str.rstrip("%").astype(float)/100
df["Risk: 18+ Estimate"]=df["Risk: 18+ Estimate"].str.rstrip("%").astype(float)/100
df["Risk: 18+ 95% CI (Lower)"]=df["Risk: 18+ 95% CI (Lower)"].str.rstrip("%").astype(float)/100
df["Risk: 18+ 95% CI (Upper)"]=df["Risk: 18+ 95% CI (Upper)"].str.rstrip("%").astype(float)/100
df["Disorder: 18+ Estimate"]=df["Disorder: 18+ Estimate"].str.rstrip("%").astype(float)/100
df["Disorder: 18+ 95% CI (Lower)"]=df["Disorder: 18+ 95% CI (Lower)"].str.rstrip("%").astype(float)/100
df["Disorder: 18+ 95% CI (Upper)"]=df["Disorder: 18+ 95% CI (Upper)"].str.rstrip("%").astype(float)/100
df2 = df
df2
```

```python
#1) Is there a state correlation?
# Bar plot: State correlation
x_axis = df.index.values
y_axis = df["Number of Prescriptions"]
fig, ax = plt.subplots(figsize=(14,4), dpi=300)
plt.bar(x_axis, y_axis, color = "green")

plt.title("Opioid Prescriptions per State (2020)", fontdict={"color":"Blue", "fontsize":20})
plt.xlabel("States", fontsize=10, color="blue")
plt.ylabel("Counts", fontsize=10, color="blue")

plt.xticks(x_axis, rotation="vertical")
plt.xlim(-0.50, len(x_axis))
plt.ticklabel_format(style="plain", axis="y")

high = df["Number of Prescriptions"].idxmax()
h_value = df["Number of Prescriptions"].max()
low = df["Number of Prescriptions"].idxmin()
l_value = df["Number of Prescriptions"].min()

print(f"States with the highest number of prescriptions in 2020 is {high} at {h_value}.")
print(f"States with the highest number of prescriptions in 2020 is {low} at {l_value}.")

# Save image
plt.tight_layout()
plt.savefig("Images/prescriptions_state.png", bbox_inches="tight")

plt.show()
```

```python
#6) Is there an age correlation? 
# Create DataFrame
line_df = pd.DataFrame(df2, columns=["Number of Prescriptions", "Total Population", "18+ Population"])
minors = df2["Total Population"] - df2["18+ Population"]
line_df.insert(3, "Minors", minors)

x = line_df.index.values
adults = line_df["18+ Population"]
minors = line_df["Minors"]

plt.figure(figsize=(14,4), dpi=700)
plt.plot(x, adults, label="Adults", marker=".", markersize=10, markeredgecolor="black")
plt.plot(x, minors, label="Minors", marker=".", markersize=10, markeredgecolor="black")

plt.title("Adults & Minors", fontdict={"color":"Blue", "fontsize":20})
plt.xlabel("States", fontsize=10, color="blue")
plt.ylabel("Counts", fontsize=10, color="blue")
plt.ticklabel_format(style="plain", axis="y")
plt.xticks(x, rotation="vertical")
plt.legend()

high = line_df["Number of Prescriptions"].idxmax()
h_value = df["Number of Prescriptions"].max()
low = df["Number of Prescriptions"].idxmin()
l_value = df["Number of Prescriptions"].min()

print(f"States with the highest number of prescriptions for adults is {high} at {h_value}.")
print(f"States with the highest number of prescriptions for minors is {low} at {l_value}.")

# Save image
plt.tight_layout()
plt.savefig("Images/adults_minors.png", bbox_inches="tight")

plt.show()
```

# Sandra

```python
# Create DataFrame, add  "RX Rate by Population" column
rx_rate = (df2["Number of Prescriptions"]/df2["Total Population"])
df2.insert(2, "RX Rate by Population", rx_rate)
pd.set_option("display.max_columns", 23, "display.colheader_justify", "center")
rx_df = df2.sort_values(by="RX Rate by Population", ascending=False)
rx_df.head()
```

```python
# Set x axis for RX Rate Analyses
x_axis_data = rx_df['RX Rate by Population']
#Set y-axis for Disorder Analyses
y_axis_data = rx_df['Disorder: 18+ Estimate']

#Plot RX Rate by Disorder
plt.scatter(x_axis_data, y_axis_data, c='green')
z = np.polyfit(x_axis_data, y_axis_data, 1)
p = np.poly1d(z)
plt.plot(x_axis_data,p(x_axis_data),"b--")
plt.title("Rates of Opiate Prescriptions by Self Reported Substance Use Disorder", fontdict={"color":"Blue", "fontsize":14})
plt.xlabel('Rate of Opiate Presciptions by Population', color="blue", fontsize=10)
plt.ylabel('Substance Use Disorder in Past Year', color="blue", fontsize=10)

# Calculate the correlation coefficient
print(f"The correlation coefficient between opiate prescribing rates and substance use disorder is {round(st.pearsonr(x_axis_data, y_axis_data)[0],2)}")

# Save image
plt.tight_layout()
plt.savefig("Images/rx_usage.png", bbox_inches="tight")

plt.show()
```

```python
# Set x axis for Risk Analyses
x_risk_data = rx_df['Risk: 18+ Estimate']
y_axis_data = rx_df['Disorder: 18+ Estimate']

#Plot Risk by Disorder

z = np.polyfit(x_risk_data, y_axis_data, 1)
p = np.poly1d(z)
plt.scatter(x_risk_data, y_axis_data, c='grey')
plt.plot(x_risk_data,p(x_risk_data),"b--")
plt.title("Reported Risk Perception by Self Reported Substance Use Disorder", fontdict={"color":"Blue", "fontsize":14})
plt.xlabel('Reported Risk Perception', color="blue", fontsize=10)
plt.ylabel('Substance Use Disorder in Past Year', color="blue", fontsize=10)
plt.xlim(0.75,0.95)
plt.ylim(0.1,0.25)

# Calculate the correlation coefficient between perceptions of risk and substance use disorder
print(f"The correlation coefficient between perceptions of risk of using heroin once or twice and substance use disorder is {round(st.pearsonr(x_risk_data, y_axis_data)[0],2)}")

# Save image
plt.tight_layout()
plt.savefig("Images/risk_usage.png", bbox_inches="tight")

plt.show()
```

```python
# Set x axis for Use Analyses
x_use_data = rx_df['Usage: 18+ Estimate']

#Plot Use by Disorder
plt.scatter(x_use_data, y_axis_data, c='red')
z = np.polyfit(x_use_data, y_axis_data, 1)
p = np.poly1d(z)
plt.plot(x_use_data,p(x_use_data),"b--")
plt.title("Reported Heroin Use by Substance Use Disorder", fontdict={"color":"Blue", "fontsize":16})
plt.xlabel('Reported Heroin Use (%)', color="blue", fontsize=10)
plt.ylabel('Substance Use Disorder in Past Year', color="blue", fontsize=10)
#plt.xlim(0,1.5)
#plt.ylim(10,25)

print(f"The correlation coefficient between reported heroin use and substance use disorder is {round(st.pearsonr(x_use_data, y_axis_data)[0],2)}")

# Save image
plt.tight_layout()
plt.savefig("Images/disorder_usage.png", bbox_inches="tight")

plt.show()
```
